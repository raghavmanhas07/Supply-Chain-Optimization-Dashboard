```python
# Calculating Exact Delay
df['Delay_Days'] = df['Days for shipping (real)'] - df['Days for shipment (scheduled)']
```

```python
df['Delay_Days'].head(10)
```

```python
# Cleaning negative delays
df['Delay_Days'] = df['Delay_Days'].apply(lambda x : x if x > 0 else 0)
```

```python
df['Delay_Days'].head(10)
```

```python
# Creating penalty cost logic
df['Late_Penalty_Cost'] = np.where(df['Delay_Days'] > 0, 10 + (df['Delay_Days'] * 5), 0)
print(df[['Delay_Days', 'Late_Penalty_Cost']].head(10))
```

```python
df['Delivery_Performance'] = np.where(df['Delay_Days'] == 0 , 'On Time', 'Late')
print(df[['Delay_Days', 'Late_Penalty_Cost', 'Delivery_Performance']].head(10))
```

```python
cols_to_keep = ['order date (DateOrders)', 'Shipping Mode', 'Customer City', 'Customer Country', 
                'Order Region', 'Category Name', 'Late_Penalty_Cost', 
                'Delivery_Performance', 'Order Item Quantity', 'Sales']
clean_df = df[cols_to_keep]
```

```python
clean_df.to_csv('Cleaned_Supply_Chain.csv', index = False)
```
